```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import json
import statistics
import math
import re
from collections import Counter
import numpy as np
import matplotlib.pyplot as plt
```

```python
file_name = "simulation_op/simulation_output_jmdrive_yellow.json"

with open(file_name, "r") as json_file:
    data = json.load(json_file)
    
keys = list(data.keys())
total_collisions = [data[key]['total collisions'] for key in keys]
emergency_brakes = [data[key]['emergency brakes'] for key in keys]

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(18, 6))

# subplot for "Total Collisions"
ax1.plot(keys, total_collisions, marker='o', label='Total Collisions', linestyle='-', color='blue')
ax1.set_xlabel('Keys')
ax1.set_ylabel('Total Collisions')
ax1.set_title('Line Plot for Total Collisions')
ax1.grid(True)
ax1.legend()

# subplot for "Emergency Brakes"
ax2.plot(keys, emergency_brakes, marker='o', label='Emergency Brakes', linestyle='-', color='red')
ax2.set_xlabel('Keys')
ax2.set_ylabel('Emergency Brakes')
ax2.set_title('Line Plot for Emergency Brakes')
ax2.grid(True)
ax2.legend()

ax1.set_xticklabels(keys, rotation=45, ha="right")
ax2.set_xticklabels(keys, rotation=45, ha="right")
plt.tight_layout()
plt.show()
```
