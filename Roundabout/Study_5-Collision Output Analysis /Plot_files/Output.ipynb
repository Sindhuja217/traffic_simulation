```python
from PIL import Image
import matplotlib.pyplot as plt
import pandas as pd

# Define the initial x and y coordinates
x = -500  # Initial x-coordinate
y = 0  # Initial y-coordinate

# Load the image
image_path = "/Users/sindhujach/Desktop/ECE_699/Traffic_simulation/Weeklytask-26-10-23/Roundabout.png"
image = Image.open(image_path)

# Set the figure size (width, height) in inches
fig, ax = plt.subplots(figsize=(16, 16))  # Adjust the width and height as needed

# Display the image with specified coordinates
ax.imshow(image, extent=[x, x + 800, y - 110,y + 310])

# Set axis limits (adjust as needed)
ax.set_xlim(x, x + 800)
ax.set_ylim(y - 110,y + 310)

# Load your collision data from the Excel file into a DataFrame
collision_data = pd.read_excel("/Users/sindhujach/Desktop/ECE_699/Traffic_simulation/Weeklytask-26-10-23/merged_data.xlsx")

# Extract 'x' and 'y' data into separate lists
x_values = collision_data['x']
y_values = collision_data['y']

# Create a hexbin plot with a minimum count threshold
hb = ax.hexbin(x_values, y_values, gridsize=50, bins='log', cmap='viridis_r', mincnt=1)



# Add a colorbar for the hexbin plot
colorbar = plt.colorbar(hb, shrink=0.3)  # Adjust the shrink value as needed
# Save the figure as an image file (e.g., PNG)
output_file_path = "/Users/sindhujach/Desktop/ECE_699/Traffic_simulation/Weeklytask-26-10-23/heatmap.png"
plt.savefig(output_file_path, dpi=300, bbox_inches='tight')

# Show the image
plt.show()
```
