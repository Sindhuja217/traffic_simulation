# Using AutoEncoders

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_score
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import matplotlib.pyplot as plt



# Standardize the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(result)

# Define the autoencoder model
input_dim = 38
encoding_dim = 2  # Choose a smaller dimension for encoding
input_layer = keras.Input(shape=(input_dim,))
encoder = layers.Dense(encoding_dim, activation='relu')(input_layer)
decoder = layers.Dense(input_dim, activation='sigmoid')(encoder)
autoencoder = keras.Model(inputs=input_layer, outputs=decoder)

# Compile the model
autoencoder.compile(optimizer='adam', loss='mse')

# Train the autoencoder
autoencoder.fit(scaled_data, scaled_data, epochs=10, batch_size=64, shuffle=True, validation_split=0.2)

# Obtain the learned representations (encoded data)
encoded_data = autoencoder.predict(scaled_data)


encoded_data
```

```python
# Apply K-means clustering
kmeans = KMeans(init= 'k-means++', max_iter= 300, n_clusters= 5, random_state= 42)
clusters = kmeans.fit_predict(encoded_data)
df_kmeans = result.copy()
df_kmeans['cluster'] = kmeans.labels_

print(df_kmeans['cluster'].value_counts())
```

```python
from sklearn import metrics
silhouette_score = metrics.silhouette_score(encoded_data, df_kmeans['cluster'], metric='euclidean')
print(silhouette_score)
```
