```python
import pandas as pd
import os

# Define the folder where your Excel files are located
folder_path = "C:/Users/saiko/OneDrive/Desktop/699/Task-3/Task-4/collisionMinGapFactor"

# Create an empty list to store the DataFrames
dfs = []

# Function to process each Excel file and create DataFrames
def process_excel_file(file_path):
    df = pd.read_excel(file_path, header=0)
    dfs.append(df)

# Iterate through files in the folder and create DataFrames
for idx, filename in enumerate(os.listdir(folder_path)):
    if filename.endswith(".xlsx"):
        file_path = os.path.join(folder_path, filename)
        process_excel_file(file_path)
        # Assign the DataFrame to the global scope with the same name
        globals()[f'df_{idx+1}'] = dfs[idx]


# Calculate the mean, variance, range, and standard deviation of each DataFrame
mean_data = pd.concat([df.mean().to_frame().T for df in dfs], ignore_index=True)
variance_data = pd.concat([df.var().to_frame().T for df in dfs], ignore_index=True)
range_data = pd.concat([(df.max() - df.min()).to_frame().T for df in dfs], ignore_index=True)
std_deviation_data = pd.concat([df.std().to_frame().T for df in dfs], ignore_index=True)

# Save the statistics to separate Excel files
mean_transposed_file_path = os.path.join(folder_path, "mean_statistics_transposed.xlsx")
variance_transposed_file_path = os.path.join(folder_path, "variance_statistics_transposed.xlsx")
range_transposed_file_path = os.path.join(folder_path, "range_statistics_transposed.xlsx")
std_deviation_transposed_file_path = os.path.join(folder_path, "std_deviation_statistics_transposed.xlsx")

mean_data.to_excel(mean_transposed_file_path, index=False, engine='openpyxl', sheet_name='Mean')
variance_data.to_excel(variance_transposed_file_path, index=False, engine='openpyxl', sheet_name='Variance')
range_data.to_excel(range_transposed_file_path, index=False, engine='openpyxl', sheet_name='Range')
std_deviation_data.to_excel(std_deviation_transposed_file_path, index=False, engine='openpyxl', sheet_name='StandardDeviation')

print("Statistics transposed and saved to Excel files.")
```

```python
import os
import pandas as pd
import matplotlib.pyplot as plt

# Define the folder paths for mean and variance data
mean_folder_path = "C:/Users/saiko/OneDrive/Desktop/699/Task-3/Task-4/Tau/mean_statistics_transposed.xlsx"
variance_folder_path = "C:/Users/saiko/OneDrive/Desktop/699/Task-3/Task-4/Tau/variance_statistics_transposed.xlsx"

# Create empty lists to store data
mean_values = []
variance_values = []

# Function to process each Excel file and extract data
def process_excel_file(file_path, data_list):
    df = pd.read_excel(file_path)
    data_list.append(df["Value"].values[0])  # Replace "Value" with the actual column name

# Process mean data files
for filename in os.listdir(mean_folder_path):
    if filename.endswith(".xlsx"):
        file_path = os.path.join(mean_folder_path, filename)
        process_excel_file(file_path, mean_values)

# Process variance data files
for filename in os.listdir(variance_folder_path):
    if filename.endswith(".xlsx"):
        file_path = os.path.join(variance_folder_path, filename)
        process_excel_file(file_path, variance_values)

# Create box plots
plt.boxplot(mean_values, vert=False, labels=mean_values, whis='range', showmeans=True, meanline=True, showcaps=True, boxprops=dict(color='b'))
plt.xlabel('Value')
plt.ylabel('Mean')
plt.title('Box Plot of Mean with Variance as Error Bars')
plt.grid(True)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Replace these file paths with the paths to your Excel files
mean_data_file = "C:/Users/saiko/OneDrive/Desktop/699/Task-3/Task-4/sigmaerror_mee/mean_statistics_transposed.xlsx"
variance_data_file = "C:/Users/saiko/OneDrive/Desktop/699/Task-3/Task-4/sigmaerror_mee/variance_statistics_transposed.xlsx"

# Read the mean and variance data from the Excel files
mean_data = pd.read_excel(mean_data_file)
variance_data = pd.read_excel(variance_data_file)

# Extract values and mean values from the data
values = mean_data['Value']
mean_values = mean_data['Collsions']

# Extract variance values
variance_values = variance_data['Collsions']

# Create aligned x positions
x_positions = np.arange(len(values))

# Create a figure
plt.figure(figsize=(10, 6))

# Create a bar plot
plt.bar(x_positions, mean_values, label='Mean', color='blue', alpha=0.6)

# Create an error plot with aligned x positions
plt.errorbar(x_positions, mean_values, yerr=np.sqrt(variance_values), fmt='o', color='red', ecolor='black', elinewidth=1.5, label='Variance')

# Set x-tick labels
plt.xticks(x_positions, values)

# Set labels and title
plt.ylabel('Mean Values')
plt.xlabel('Parameter Values')
plt.title('Bar Plot for sigmaerror with Error Bars (Variance)')

# Add a legend
plt.legend()

# Add a grid
plt.grid()

# Show the plot
plt.show()
```
