```python
import pandas as pd
import numpy as np
# Example DataFrame
# Assuming your data is in a DataFrame named 'df'

# Group the DataFrame by 'time' and 'lane'
grouped = fcd_df.groupby(['time', 'lane'])

# Create empty columns to store nearest vehicle ID and least distance
fcd_df['nearest_vehicle_id'] = "0"
fcd_df['least_distance'] = 0

for name, group in grouped:
    for index, row in group.iterrows():
        id_pos = row['pos']
        other_ids = group[group['id'] != row['id']]  # Excluding the same id from comparison
        other_ids['distance_to_' + str(row['id'])] = abs(other_ids['pos'] - id_pos)
        
        if len(other_ids) > 0:
            min_distance_row = other_ids.loc[other_ids['distance_to_' + str(row['id'])].idxmin()]
            fcd_df.at[index, 'nearest_vehicle_id'] = min_distance_row['id']
            fcd_df.at[index, 'least_distance'] = min_distance_row['distance_to_' + str(row['id'])]

# Display the updated DataFrame with the columns identifying the nearest 'id' and its least distance
print(fcd_df)
```

```python
import numpy as np
from keras.models import Sequential
from keras.layers import LSTM, Dense
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.utils import to_categorical

# Assuming X_train and y_train are already defined and properly scaled

# Encode the labels to integers
label_encoder = LabelEncoder()
encoded_y_train = label_encoder.fit_transform(y_train)
encoded_y_test = label_encoder.transform(y_test)

# Convert integers to dummy variables (i.e., one-hot encoded)
dummy_y_train = to_categorical(encoded_y_train)
dummy_y_test = to_categorical(encoded_y_test)

# Ensure labels are of numeric data type (float32)
dummy_y_train = dummy_y_train.astype('float32')
dummy_y_test = dummy_y_test.astype('float32')


# Reshape input to be [samples, time steps, features] which is required for LSTM layer
# Since your data doesn't have a time step, we can treat each row as 1 time step.
X_train_reshaped = np.reshape(X_train, (X_train.shape[0], 1, X_train.shape[1]))
X_test_reshaped = np.reshape(X_test, (X_test.shape[0], 1, X_test.shape[1]))
```

```python
# Convert data types if needed
X_train_reshaped = np.asarray(X_train_reshaped, dtype=np.float32)
dummy_y_train = np.asarray(dummy_y_train, dtype=np.float32)
X_test_reshaped = np.asarray(X_test_reshaped, dtype=np.float32)
dummy_y_test = np.asarray(dummy_y_test, dtype=np.float32)
```

```python
from tensorflow.keras.optimizers import Adam

# Define LSTM model
model = Sequential()
model.add(LSTM(50, input_shape=(X_train_reshaped.shape[1], X_train_reshaped.shape[2])))
model.add(Dense(dummy_y_train.shape[1], activation='softmax'))

# Compile the model
model.compile(loss='categorical_crossentropy', optimizer=Adam(), metrics=['accuracy'])

# Fit the model
model.fit(X_train_reshaped, dummy_y_train, epochs=10, batch_size=32, validation_data=(X_test_reshaped, dummy_y_test), verbose=2)

# # Evaluate the model
# scores = model.evaluate(X_test_reshaped, dummy_y_test, verbose=0)
# print("Accuracy: %.2f%%" % (scores[1]*100))
```

```python
from sklearn.metrics import classification_report

# Generate predictions on the test set
y_pred = model.predict(X_test_reshaped)

# Convert one-hot encoded predictions to class labels
y_pred_classes = np.argmax(y_pred, axis=1)
y_true_classes = np.argmax(dummy_y_test, axis=1)

# Print classification report
print("Classification Report:\n", classification_report(y_true_classes, y_pred_classes))
```

```python
import xml.etree.ElementTree as ET
import pandas as pd

# Assuming 'xml_data' contains the XML data you provided
# root = ET.fromstring("C:/Users/saiko/OneDrive/Desktop/699/Task-6/full-output.xml")
tree = ET.parse("C:/Users/saiko/OneDrive/Desktop/699/Task-6/full-output.xml")
root = tree.getroot()


# Create an empty dictionary to store vehicle counts for each lane at each timestamp
lane_vehicle_counts = {}

# Iterate through 'data' tags in the XML
for data in root.iter('data'):
    timestamp = data.get('timestep')  # Get the timestamp
    lane_vehicle_counts[timestamp] = {}
    
    # Iterate through 'lane' tags to extract vehicle counts for each lane
    for lane in data.iter('lane'):
        lane_id = lane.get('id')  # Get the lane ID
        vehicle_count = int(lane.get('vehicle_count')) if 'vehicle_count' in lane.attrib else 0
        lane_vehicle_counts[timestamp][lane_id] = vehicle_count

# Convert the dictionary to a DataFrame
lane_vehicle_counts_df = pd.DataFrame(lane_vehicle_counts).fillna(0).astype(int)
```
