```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import gaussian_kde

# Define the initial x and y coordinates
x = -100  # Initial x-coordinate
y = 0  # Initial y-coordinate

# Load the image
image_path = "C:/Users/saiko/OneDrive/Desktop/699/Task-3/demonew.png"
image = Image.open(image_path)

# Load your collision data from the Excel file into a DataFrame
collision_data = pd.read_excel("C:/Users/saiko/OneDrive/Desktop/699/Task-5/merged_data.xlsx")

# Extract 'x' and 'y' data into separate lists
x_values = collision_data['x']
y_values = collision_data['y']

# Calculate density and create a contour plot
def calculate_density(x_values, y_values, grid_resolution=100):
    points = np.vstack([x_values, y_values])
    kde = gaussian_kde(points)

    x, y = np.meshgrid(
        np.linspace(min(x_values), max(x_values), grid_resolution),
        np.linspace(min(y_values), max(y_values), grid_resolution)
    )

    grid_points = np.vstack([x.ravel(), y.ravel()])
    density = kde(grid_points).reshape(x.shape)

    return x, y, density

x_grid, y_grid, density_values = calculate_density(x_values, y_values, grid_resolution=100)

# Set the figure size for the image and density plot
fig, ax = plt.subplots(figsize=(12, 12))  # Adjust the width and height as needed

# Display the image with specified coordinates and extent
extent = [x, x + 500, y - 120, y + 120]
ax.imshow(image, extent=extent)

# Set axis limits (adjust as needed)
ax.set_xlim(x, x + 500)
ax.set_ylim(y - 120, y + 120)

CS = ax.contour(x_grid, y_grid, density_values,cmap='hsv', extend='both',
                linewidths=2, extent=(-3, 3, -2, 2),vmin=0, vmax=0.000105)
ax.clabel(CS, inline=True, fontsize=10)
ax.set_title('Simplest default with labels')
# make a colorbar for the contour lines
CB = fig.colorbar(CS, shrink=0.5)

# # Create a contour plot using density values with the same extent as the image
# plt.contourf(x_grid, y_grid, density_values, levels=20, cmap='viridis', alpha=0.5, extent=extent)
# plt.colorbar(label='Density')
# plt.xlabel('X')
# plt.ylabel('Y')
# plt.title('Density Contour Plot')
# Save the figure as an image file (e.g., PNG)
output_file_path = "C:/Users/saiko/OneDrive/Desktop/699/Task-5/heatmap4.png"
plt.savefig(output_file_path, dpi=300, bbox_inches='tight')
# Show the plot
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the collision data from an Excel file
collision_data = pd.read_excel("C:/Users/saiko/OneDrive/Desktop/Collision_data.xlsx")
collision_data

# Calculate the minimum and maximum positions
min_pos = collision_data['pos'].min()
max_pos = collision_data['pos'].max()

# Define the bin width
bin_width = 10  # You can adjust this as needed

# Create position ranges with fixed intervals
position_ranges = [i for i in range(int(min_pos), int(max_pos) + bin_width, bin_width)]

# Function to map position to range
def map_to_range(pos):
    for i in range(len(position_ranges) - 1):
        if position_ranges[i] <= pos < position_ranges[i + 1]:
            return f"{position_ranges[i]}-{position_ranges[i + 1]}"
    return f"{position_ranges[-1]}+"

# Add the position_ranges as a new column in the DataFrame
collision_data['position_range'] = collision_data['pos'].map(map_to_range)

# Sort the position ranges correctly
sorted_ranges = sorted(collision_data['position_range'].unique(), key=lambda x: int(x.split('-')[0]))

# Plot the counts of collisions in each position range with sorted ranges
plt.figure(figsize=(12, 6))
sns.countplot(data=collision_data, x='position_range', hue='type', order=sorted_ranges, palette='viridis')
plt.xlabel('Position Range')
plt.ylabel('Count')
plt.title('Collisions by Position Range')
plt.xticks(rotation=45)
plt.legend(title='Collision Type', loc='upper right')
plt.show()
```
